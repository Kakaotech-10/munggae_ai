* unsmile 데이터에서 '기타 혐오'와 가장 가까운 성격의 칼럼은 '악플/욕설'이라고 판단.
* 처리 후 '기타 혐오' 칼럼 삭제

```python
unsmile
```

```python
hatescore_bad_others['microlabel'] = hatescore_bad_others['microlabel'].apply(lambda x: x.split(','))
```

```python
hatescore_bad_others['microlabel'] = hatescore_bad_others['microlabel'].apply(lambda x: [label.strip() for label in x])
```

* hatescore_bad_others의 microlabel의 속성이 2개 이상 있을 경우, 이 속성을 추후 unsmile 데이터에 반영하기 위해 문자열 형태에서 리스트 형태로 변경.

```python
hatescore_bad_others = hatescore_bad_others[['comment', 'microlabel']]
```

```python
hatescore_bad_others
```

```python
hatescore_bad_others['여성/가족'] = 0
hatescore_bad_others['남성'] = 0
hatescore_bad_others['성소수자'] = 0
hatescore_bad_others['인종/국적'] = 0
hatescore_bad_others['연령'] = 0
hatescore_bad_others['지역'] = 0
hatescore_bad_others['종교'] = 0
hatescore_bad_others['기타 혐오'] = 0
hatescore_bad_others['악플/욕설'] = 0
hatescore_bad_others['clean'] = 0
```

```python
hatescore_bad_others
```

* hatescore_bad_others과 unsmile 데이터셋 형태 일치화

```python
# 대상 열 목록
target_columns = ['여성/가족', '남성', '성소수자', '인종/국적', '연령', '지역', '종교', '기타 혐오', '악플/욕설', 'clean']

# microlabel 열을 기준으로 각 열 값을 업데이트
for index, row in hatescore_bad_others.iterrows():
    for label in row['microlabel']:
        if label in target_columns:
            hatescore_bad_others.loc[index, label] = 1
```

* hatescore_bad_others의 microlabel의 속성을 unsmile 형식의 데이터셋에 반영.

```python
hatescore_bad_others
```

```python
# microlabel 열에서 리스트 길이가 2 이상인 행 필터링
filtered_rows = hatescore_bad_others[hatescore_bad_others['microlabel'].apply(len) >= 2]

# 결과 출력
filtered_rows
```

* microlabel의 속성이 2개 이상일 경우에도 데이터셋에 잘 반영됨.

```python
hatescore_bad_others.drop(columns="microlabel", axis=1, inplace=True)
hatescore_bad_others.rename(columns={'comment': '문장'}, inplace=True)
hatescore_bad_others.loc[hatescore_bad_others['기타 혐오'] == 1, '악플/욕설'] = 1
hatescore_bad_others.drop(columns='기타 혐오', axis=1, inplace=True)
```

```python
hatescore_bad_others
```

* unsmile 데이터셋 형태와 일치화 작업

```python
unsmile = pd.concat([unsmile, hatescore_bad_others], ignore_index=True)
```

```python
unsmile
```

* unsmile이라는 하나의 데이터셋으로 모든 데이터 결합 완료.
